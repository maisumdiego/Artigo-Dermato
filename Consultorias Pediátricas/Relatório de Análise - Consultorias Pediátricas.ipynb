```python
import pandas as pd
import seaborn as sns
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
dados = pd.read_excel(r'C:\Users\Diego\Projetos Python\Artigo Dermato\Planilha_tratada.xlsx', index_col=False)
```

```python
hosp_pediatrico = dados[dados.Hospital == 6]
```

# Classificação de faixa etária

```python
for i in range(len(hosp_pediatrico['Idade (meses)'])):
    if hosp_pediatrico['Idade (meses)'].iloc[i] <= 1:
        hosp_pediatrico['Idade (anos)'].iloc[i] = (hosp_pediatrico['Idade (meses)'].iloc[i]/12)
    elif hosp_pediatrico['Idade (meses)'].iloc[i] > 1:
        hosp_pediatrico['Idade (anos)'].iloc[i] = (hosp_pediatrico['Idade (meses)'].iloc[i]/12)
```

```python
classes = [0, 0.084, 0.91666667 , 2, 5, 10, 17]
labels = ['0 - 1 mês', '2 - 11 meses' , '1 - 2 anos', '3 - 5 anos', '6 - 10 anos', '11 - 17 anos']

hosp_pediatrico['Faixa Etária'] = pd.cut(x= hosp_pediatrico['Idade (anos)'], 
       bins=classes, 
       labels=labels, 
       include_lowest=True)
```

```python
freq_etaria = hosp_pediatrico.groupby('Faixa Etária')[['Nº Requisição']].count().rename(
    columns={'Nº Requisição':'Frequência'})
freq_etaria['%'] = ((freq_etaria['Frequência'] / freq_etaria['Frequência'].sum()) * 100).round(2)
```

```python
freq_etaria
```

```python
freq_etaria.to_excel('Frequência Consultorias por Faixa Etária.xlsx', index=False)
```

```python
classe1 = hosp_pediatrico[['Faixa Etária', 'Classe de doença (CID1)', 'CID 1']]
classe1.rename(columns={'Classe de doença (CID1)':'Classe', 'CID 1':'CID'}, inplace=True)
classe2 = hosp_pediatrico[['Faixa Etária', 'Classe de doença (CID2)', 'CID 2']].dropna()
classe2.rename(columns={'Classe de doença (CID2)':'Classe', 'CID 2':'CID'}, inplace=True)
classe3 = hosp_pediatrico[['Faixa Etária', 'Classe de doença (CID3)', 'CID 3']].dropna()
classe3.rename(columns={'Classe de doença (CID3)':'Classe', 'CID 3':'CID'}, inplace=True)
```

```python
classe_cid_faixa = pd.concat([classe1, classe2, classe3], ignore_index=True)
classe_cid_faixa
```

```python
classe_por_faixa = classe_cid_faixa.groupby('Faixa Etária')['Classe'].value_counts().to_frame()
classe_por_faixa.rename(columns={'Classe':'Frequência'}, inplace= True)
```

# Classes por Faixa Etária

```python
classe_por_faixa = classe_cid_faixa.groupby('Faixa Etária')['Classe'].value_counts().to_frame()
classe_por_faixa.rename(columns={'Classe':'Frequência'}, inplace= True)
classe_por_faixa.reset_index(inplace=True)
```

```python
porcentagem_relativa = []
for x in range(classe_por_faixa.shape[0]):
    y = classe_por_faixa.iloc[x][0]
    n = ((classe_por_faixa.loc[x][2] / classe_por_faixa[classe_por_faixa['Faixa Etária'] == y]['Frequência'].sum()) * 100).round(2)
    porcentagem_relativa.append(n)  

classe_por_faixa['% Relativa'] = porcentagem_relativa
```

```python
classe_por_faixa['% Geral'] = ((classe_por_faixa['Frequência'] / classe_por_faixa['Frequência'].sum()) * 100).round(2)
classe_por_faixa
```

```python
classe_por_faixa.to_excel('Classes por Faixa Etária.xlsx', index=False)
```

# CIDs por Faixa Etária

```python
cid_por_faixa = classe_cid_faixa.groupby('Faixa Etária')['CID'].value_counts().to_frame()
cid_por_faixa.rename(columns={'CID':'Frequência'}, inplace= True)
cid_por_faixa.reset_index(inplace=True)
```

```python
porcentagem_relativa = []
for x in range(cid_por_faixa.shape[0]):
    y = cid_por_faixa.iloc[x][0]
    n = ((cid_por_faixa.loc[x][2] / cid_por_faixa[cid_por_faixa['Faixa Etária'] == y]['Frequência'].sum()) * 100).round(2)
    porcentagem_relativa.append(n)  

cid_por_faixa['% Relativa'] = porcentagem_relativa
```

```python
cid_por_faixa['% Geral'] = ((cid_por_faixa['Frequência'] / cid_por_faixa['Frequência'].sum()) * 100).round(2)
```

```python
tabela_final_top5 = pd.DataFrame()
for x in cid_por_faixa['Faixa Etária'].unique():
        n = cid_por_faixa[cid_por_faixa['Faixa Etária'] == x].head()
        tabela_final_top5 = tabela_final_top5.append(n)
```

```python
tabela_final_top5.reset_index(drop=True, inplace=True)
tabela_final_top5.index = range(1, tabela_final_top5.shape[0] + 1)
tabela_final_top5
```

```python
tabela_final_top5.to_excel('Top 5 CIDs por Faixa Etária.xlsx', index=False)
```
